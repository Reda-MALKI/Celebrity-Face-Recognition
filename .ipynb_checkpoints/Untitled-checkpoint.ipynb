```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
from skimage.feature import hog
```

```python
image1 = cv2.imread("Sports-celebrity-images/Kobe-Bryant/kobe_(23).jpg")
image1 = cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,6))
plt.imshow(image1,cmap="gray")
plt.axis("off")

plt.show()
```

```python
data_dir = "Sports-celebrity-images"
images = []
labels =  []

labels_names = os.listdir(data_dir)
for label in labels_name:
    folder = os.path.join(data_dir,label)
    for img_file in os.listdir(folder):
        img_path = os.path.join(folder,img_file)
        img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img,(128,128))
        features,_ = hog(img,orientations = 9 , pixel_per_cell=(8,8),cells_per_block = (2,2),
                         visualize = True,multichannel = False)
        images.append(features)
        labels.append(label)

x = np.array(images)
y = np.array(lables)
        
```
