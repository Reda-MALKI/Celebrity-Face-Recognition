#### This code loads images of sports celebrities from labeled folders, detects faces using a Haar cascade, and extracts features from the detected face regions. It computes HOG features on grayscale resized faces and color histograms on resized color faces, then combines these features into a single vector per image. Finally, it collects all feature vectors and their corresponding labels into arrays for further use in a machine learning task.

```python
import os
import cv2
import numpy as np
from skimage.feature import hog

data_dir = "Sports-celebrity-images"
images = []
labels = []


face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

labels_names = os.listdir(data_dir)
print(labels_names)

for label in labels_names:
    folder = os.path.join(data_dir, label)
    for img_file in os.listdir(folder):
        img_path = os.path.join(folder, img_file)
        
        img_color = cv2.imread(img_path)
        if img_color is None:
            continue  # Skip unreadable files

        img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

        
        faces = face_cascade.detectMultiScale(img_gray, scaleFactor=1.1, minNeighbors=5)
        if len(faces) == 0:
            continue  # Skip if no face detected

        
        x, y, w, h = faces[0]
        face_region = img_gray[y:y+h, x:x+w]
        face_color_region = img_color[y:y+h, x:x+w]

        
        face_resized_gray = cv2.resize(face_region, (128, 128))
        face_resized_color = cv2.resize(face_color_region, (64, 64))

        
        features, _ = hog(face_resized_gray, orientations=9, pixels_per_cell=(8, 8),
                          cells_per_block=(2, 2), visualize=True)

        
        hist_b = cv2.calcHist([face_resized_color], [0], None, [32], [0, 256]).flatten()
        hist_g = cv2.calcHist([face_resized_color], [1], None, [32], [0, 256]).flatten()
        hist_r = cv2.calcHist([face_resized_color], [2], None, [32], [0, 256]).flatten()

       
        full_feature = np.concatenate([features, hist_b, hist_g, hist_r])
        images.append(full_feature)
        labels.append(label)

x = np.array(images)
y = np.array(labels)
```

#### an overall example , a function that applies all the technics that i've fount the best for the model and try a prediction the results are correct 

```python
def detect_face(img_path):
    import joblib
    import cv2
    import numpy as np
    from skimage.feature import hog

    face_cascade = cv2.CascadeClassifier(
        cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
    )

    image_color = cv2.imread(img_path)
    if image_color is None:
        raise ValueError("Image not found or path is incorrect.")

    image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(image_gray, 1.1, 5)
    if len(faces) == 0:
        raise ValueError("No face detected in the image.")

    x, y, w, h = faces[0]
    face_color = image_color[y:y+h, x:x+w]
    face_gray = image_gray[y:y+h, x:x+w]

    face_resized_gray = cv2.resize(face_gray, (128, 128))
    face_resized_color = cv2.resize(face_color, (64, 64))

    features, _ = hog(face_resized_gray, orientations=9, pixels_per_cell=(8, 8),
                      cells_per_block=(2, 2), visualize=True)

    hist_b = cv2.calcHist([face_resized_color], [0], None, [32], [0, 256]).flatten()
    hist_g = cv2.calcHist([face_resized_color], [1], None, [32], [0, 256]).flatten()
    hist_r = cv2.calcHist([face_resized_color], [2], None, [32], [0, 256]).flatten()

    full_features = np.concatenate([features, hist_b, hist_g, hist_r])
    x = np.array(full_features).reshape(1, -1)

    #  Load scaler and model
    scaler = joblib.load('scaler.pkl')
    model_svm = joblib.load('model_svm.pkl')

    x_scaled = scaler.transform(x)

    return model_xgb.predict(x_scaled)[0]
# Example usage
print(detect_face("Sports-celebrity-images/Kobe-Bryant/kobe_(49).jpg"))
print(detect_face("Sports-celebrity-images/Ronaldo/ronaldo_(10).jpg"))
```
